# Exploratory Data Analysis using pandas_profiling, matplotlib, and seaborn

### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_profiling import ProfileReport
```

### Load JSON/dict as dataframe

```python
df_classification = pd.read_json('classification.json')
df_regression = pd.read_json('regression.json')
df_cleaning = pd.read_csv('data/Data.csv')
```

### Use pandas_profiling to profile data quickly

```python
classification_profile = ProfileReport(df_classification, title='Classification Data Report', minimal=True)
```

```python
classification_profile
```

### Export profiling report as HTML

```python
classification_profile.to_file('classification.html')
```

```python
cleaning_profile = ProfileReport(df_cleaning, title='Cleaning Data Report', minimal=True)
```

```python
cleaning_profile
```

```python
cleaning_profile.to_file('cleaning.html')
```

### Try to visualize dirty data

```python
%matplotlib inline
```

```python
df_cleaning.plot()
plt.show()
```

### Visualize clean data

```python
df_regression.plot()
plt.show()
```

```python
df_regression.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

```python
df_classification.hist()
plt.show()
```

### Use pairplot to identify patterns from data quickly

```python
sns.pairplot(df_regression)
plt.show()
```

```python
sns.pairplot(df_cleaning)
plt.show()
```
